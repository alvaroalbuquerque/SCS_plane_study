## Generating both planes

```python
image_05_05_9_data = pd.read_csv(f"results/salt_and_pepper/salt_and_pepper_salt_0.05_pepper_0.05.csv")
image_3_3_4_data = pd.read_csv(f"results/salt_and_pepper/salt_and_pepper_salt_0.3_pepper_0.3.csv")
```

```python
for salt_p, pepper_p, data_ss in [('0.05', '0.05', image_05_05_9_data), ('0.3', '0.3', image_3_3_4_data)]:
    # SMOOTHNESS
    plt.figure(1)
    plt.clf() # clears the figure
    plt.rcParams.update({"font.size": 14})
    scatter = plt.scatter(data_ss["Smoothness"], data_ss["Curve structure"], c=data_ss['q'], cmap='gist_rainbow')
    # Overlay special symbols for points where data_ss['q'] is 0, 1, or 2
    special_symbols = {0: 'o', 1: 's', 2: '^'}
    for q_value, marker in special_symbols.items():
        subset = data_ss[data_ss['q'] == q_value]
        plt.scatter(subset["Smoothness"], subset["Curve structure"], c='white', edgecolors='black', marker=marker, label=f'q={q_value}')

    plt.colorbar(scatter, label='Parameter q')
    # plt.ylim([0, 1])
    # plt.xlim([0, 1])
    plt.ylim([-1, 1])
    plt.xlim([-1, 1])
    plt.xlabel('Smoothness')
    plt.ylabel('Curve structure')
    plt.title(f'SC S&P salt={salt_p} pep={pepper_p}')

    plt.savefig(f'results/salt_and_pepper/planes/salt_and_pepper_SCplane_weighted_salt_and_pepper_salt_{salt_p}_pepper_{pepper_p}.png')
            
    # CECP
    plt.figure(2)
    plt.clf() # clears the figure
    plt.rcParams.update({"font.size": 17})
    scatter = plt.scatter(data_ss['Weighted Entropy Correct'], data_ss['Weighted Complexity Correct'], c=data_ss['q'], cmap='gist_rainbow')
    # Overlay special symbols for points where data_ss['q'] is 0, 1, or 2
    special_symbols = {0: 'o', 1: 's', 2: '^'}
    for q_value, marker in special_symbols.items():
        subset = data_ss[data_ss['q'] == q_value]
        plt.scatter(subset["Weighted Entropy Correct"], subset["Weighted Complexity Correct"], c='white', edgecolors='black', marker=marker, label=f'q={q_value}')
    plt.colorbar(scatter, label='Parameter q')
    # plt.ylim([0, 1])
    # plt.xlim([0, 1])
    plt.ylim([-1, 1])
    plt.xlim([-1, 1])
    plt.xlabel('Weighted Entropy')
    plt.ylabel('Weighted Complexity')
    plt.title(f'CECP S&P salt={salt_p} pep={pepper_p}')

    plt.savefig(f'results/salt_and_pepper/planes/salt_and_pepper_CECPplane_weighted_salt_and_pepper_salt_{salt_p}_pepper_{pepper_p}.png')
    plt.clf() # clears the figure
```

## Plotting the probabilities

```python
image_05_05_9_data = pd.read_csv(f"results/salt_and_pepper/salt_and_pepper_salt_0.05_pepper_0.05.csv")
image_3_3_4_data = pd.read_csv(f"results/salt_and_pepper/salt_and_pepper_salt_0.3_pepper_0.3.csv")
```

```python
data_ss = image_05_05_9_data
# # get cumulative probabilities prob0,prob1,prob2
data_ss['prob0_cumsum'] = data_ss['prob0'].cumsum()
data_ss['prob1_cumsum'] = data_ss['prob1'].cumsum()
data_ss['prob2_cumsum'] = data_ss['prob2'].cumsum()

# probabilities
markers = ['o', 's', '^', 'v', '>', '<', 'p', '*', '+', 'x']
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown']
plt.clf() # clears the figure
plt.figure(1,figsize=(20,10))
plt.rcParams.update({"font.size": 17})
filtered_data = data_ss
# filtered_data = data_ss.iloc[88:284:5]  # Select every 5th row
# filtered_data = data_ss.iloc[74:282]  # Select every 5th row
for i in range(0,3):
    plt.scatter(filtered_data['q'], filtered_data[f'prob{i}'], label=f'p_{i}',
            color=colors[i % len(colors)], marker=markers[i % len(markers)])  # Plot x vs. y
    plt.plot(filtered_data['q'], filtered_data[f'prob{i}'], linestyle='-', color=colors[i % len(colors)], linewidth=0.5)


plt.xlabel('q')  # Label for x-axis
plt.ylabel('Probabilities')  # Label for y-axis
plt.title(f'Plot of q vs Probabilities (5/5/90)')  # Title of the plot
plt.legend()  # Add a legend to distinguish the curves
plt.grid(True)  # Optionally add a grid for better readability
plt.savefig(f'results/salt_and_pepper/plots/salt_and_pepper_salt_0.05_pepper_0.05_probs.png')
    
    
```

```python
plt.clf() # clears the figure
# cummulative probabilities
for i in range(0,3):
    plt.scatter(filtered_data['q'], filtered_data[f'prob{i}_cumsum'], label=f'p_{i}',
            color=colors[i % len(colors)], marker=markers[i % len(markers)])  # Plot x vs. y
    plt.plot(filtered_data['q'], filtered_data[f'prob{i}_cumsum'], linestyle='-', color=colors[i % len(colors)], linewidth=0.5)


plt.xlabel('q')  # Label for x-axis
plt.ylabel('Cummulative Probabilities')  # Label for y-axis
plt.title(f'Plot of q vs Cummulative Probabilities')  # Title of the plot
plt.legend()  # Add a legend to distinguish the curves
plt.grid(True)  # Optionally add a grid for better readability
```
