```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
import utils.noise_generator as noise_gen
import utils.utils as utils
from PIL import Image
```

```python
# [0.0, 0.05, 0.1, 0.2, 0.3, 0.5, 0.75, 1.0, 1.5, 2.0, 2.5,
# 3.0, 4.0, 5.0, 6.0, 8.0, 10.0, 15.0, 20.0, 30.0, 50.0]

img_name = "noise_original"
# Load an image
image = Image.open(f'data/misc/{img_name}.png')

gs_image = utils.normalize_uint8_to_float64(np.array(image.convert('L')))  # Convert to grayscale
print(gs_image.dtype)
print(gs_image.min(), gs_image.max())

# Add speckle noise with default parameters
noised_image = noise_gen.add_poisson_noise(gs_image, factor=0)

# Define different noise factors
noise_factors = [0, 1, 5, 10, 20]

# Create a figure with subplots
plt.figure(figsize=(15, 10))

# Plot original image
plt.subplot(2, 3, 1)
plt.imshow(gs_image, cmap='gray')
plt.title('Original image')
plt.axis('off')

# Plot images with different noise factors
for i, factor in enumerate(noise_factors, start=2):
    noisy = noise_gen.add_poisson_noise(gs_image, factor=factor)
    plt.subplot(2, 3, i)
    plt.imshow(noisy, cmap='gray')
    plt.title(f'Poisson noise (factor={factor})')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img_name = "noise_original"
# Load an image
image = Image.open(f'data/misc/{img_name}.png')

gs_image = utils.normalize_uint8_to_float64(np.array(image.convert('L')))  # Convert to grayscale
print(gs_image.dtype)
print(gs_image.min(), gs_image.max())

for factor in [5, 20]:
    # Add speckle noise with specified factor
    noised_image = noise_gen.add_poisson_noise(gs_image, factor=factor)
    
    im = Image.fromarray(noised_image)
    
    # Convert floating point image to uint8 (0-255 range)
    if im.mode == 'F':
        # Normalize and convert to 8-bit
        im_array = np.array(im)
        im_normalized = (im_array*255).astype(np.uint8)
        im = Image.fromarray(im_normalized, mode='L')  # 'L' is 8-bit grayscale
    
    im.save(f'data/misc/{img_name}_poisson_factor{factor}.png')
```

## Generating both planes

```python
image_title = 'Marble-D58'
data_ss = pd.read_csv(f"results/poisson_noise/{image_title}.csv")
```

```python
# SMOOTHNESS
plt.figure(1)
plt.clf() # clears the figure
plt.rcParams.update({"font.size": 14})
scatter = plt.scatter(data_ss["Smoothness"], data_ss["Curve structure"], c=data_ss['q'], cmap='gist_rainbow')
# Overlay special symbols for points where data_ss['q'] is 0, 1, or 2
special_symbols = {0: 'o', 1: 's', 2: '^'}
for q_value, marker in special_symbols.items():
    subset = data_ss[data_ss['q'] == q_value]
    plt.scatter(subset["Smoothness"], subset["Curve structure"], c='white', edgecolors='black', marker=marker, label=f'q={q_value}')

plt.colorbar(scatter, label='Parameter q')
# plt.ylim([0, 1])
# plt.xlim([0, 1])
plt.ylim([-1, 1])
plt.xlim([-1, 1])
plt.grid()
plt.xlabel('Smoothness')
plt.ylabel('Curve structure')
plt.title(f'SC {image_title} - Poisson Noise')

plt.savefig(f'results/poisson_noise/planes/{image_title}_SCplane_weighted.png', bbox_inches='tight')
        
# CECP
plt.figure(2)
plt.clf() # clears the figure
plt.rcParams.update({"font.size": 17})
scatter = plt.scatter(data_ss['Weighted Entropy Correct'], data_ss['Weighted Complexity Correct'], c=data_ss['q'], cmap='gist_rainbow')
# Overlay special symbols for points where data_ss['q'] is 0, 1, or 2
special_symbols = {0: 'o', 1: 's', 2: '^'}
for q_value, marker in special_symbols.items():
    subset = data_ss[data_ss['q'] == q_value]
    plt.scatter(subset["Weighted Entropy Correct"], subset["Weighted Complexity Correct"], c='white', edgecolors='black', marker=marker, label=f'q={q_value}')
plt.colorbar(scatter, label='Parameter q')
# plt.ylim([0, 1])
# plt.xlim([0, 1])
plt.ylim([-1, 1])
plt.xlim([-1, 1])
plt.grid()
plt.xlabel('Weighted Entropy')
plt.ylabel('Weighted Complexity')
plt.title(f'CECP {image_title} - Poisson Noise')

plt.savefig(f'results/poisson_noise/planes/{image_title}_CECPplane_weighted.png', bbox_inches='tight')
plt.clf() # clears the figure
```
